**1. В колоде 52 карты. Мы решили вытащить из неё одну карту. Какова вероятность того, что это будет туз?**

```python
koloda = 52

tuz_count = 4

p_tuz = tuz_count / koloda

print(f"Вероятность вытащить туз из колоды карт: {p_tuz:.2f}")
```

**2. Какова вероятность того что у трёх случайно выбранных людей будет одинаковая группа крови? Подсказка: для решения этой задачи нужно использовать как правило умножения так и правило сложения.**

```python
p_o = 0.44  # вероятность иметь 1 группу крови 
p_a = 0.42  # вероятность иметь 2 группу крови 
p_b = 0.1   # вероятность иметь 3 группу крови 
p_ab = 0.04 # вероятность иметь 4 группу крови 

#вероятность того что у всех троих выбранных людей 1ая группа крови
p_o_o_o = p_o * p_o * p_o

#вероятность того что у всех троих выбранных людей 2ая группа крови
p_a_a_a = p_a * p_a * p_a

#вероятность того что у всех троих выбранных людей 3я группа крови
p_b_b_b = p_b * p_b * p_b

#вероятность того что у всех троих выбранных людей 4я группа крови
p_ab_ab_ab = p_ab * p_ab * p_ab

#вероятность по правилу сложения т.е вероятность того что будет одинаковая группа крови у троих

p = p_o_o_o + p_a_a_a + p_b_b_b + p_ab_ab_ab

print(f"Вероятность того что у трёх случайно выбранных людей будет одинаковая группа крови: {p:.2f}")
```

**3. Студент, сдаёт тест состоящий из 30 вопросов с четырьмя вариантами ответов, из которых лишь один правильный. Так как он вообще не посещал лекции, он просто наугад выбирает один из вариантов для каждого вопроса. Какова вероятность что он правильно ответит хотя бы на один вопрос?**

```python
import math 

#вероятность угадать правильный ответ на один вопрос 
p_right1 = 1 / 4

#вероятность ответить неправильно на один вопрос
p_notright1 = 1 - p_right1

# вероятность ответить неправильно на все 30 вопросов 
p_notright1to30 = p_notright1 ** 30

# вероятность ответить правильно хотя бы на один вопрос
p_right = 1 - p_notright1to30

print(f"вероятность ответить правильно хотя бы на один вопрос: {round(p_right, 4)}")
```

**4. Какова вероятность что хотя бы у одного из двух случайно выбранных людей месяц рождения такой же как и у вас?**

```python
# всего месяцев в году
month_counts = 12 

#вероятность у одного человека совпадение месяца со мной 
p_month_with_mine = 1 / month_counts

#вероятность у одного человека не совпадения месяца со мной 
p_month_not_with_mine = 1 - p_month_with_mine 

#вероятность того что у обоих не совбадает месяц со мной 
p_both_not_with_mine = p_month_not_with_mine ** 2

# вероятность того что хотя бы у одного совпадает со мной
p = 1 - p_both_not_with_mine

print(f"вероятность что хотя бы у одного из двух случайно выбранных людей месяц рождения такой же как и у вас: {round(p,4)}")
```

**5. В колоде 52 карты. Мы решили вытащить из неё одну карту. Какова вероятность того что это будет либо валет, либо пиковая карта?**

```python
# кол-во карт
koloda = 52

#кол-во вальтов
valet_cnt = 4

#кол-во типов фигур
type_fig_cnt = 4

# кол-во пиковых карт
pik_cnt = koloda / type_fig_cnt

#кол-во пересечений. то есть случай когда и валет и пик. пиковый валет
pik_valet = 1

# кол-во правильных событий
right_cnt = valet_cnt + pik_cnt - pik_valet

#вероятность того что вытащенная карта будет либо валет, либо пиковая карта
p = right_cnt / koloda

print(f"вероятность того что это будет либо валет, либо пиковая карта: {round(p, 4)}")
```
